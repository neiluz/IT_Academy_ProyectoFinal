## Preparación de la data de prueba (Test Data)

Para llevar a cabo en este proyecto la recolección de datos, se utilizará la librería Twint, una herramienta de web scrapping de Twitter que permite obtener tweets públicos sin necesidad de una API. La ventaja de utilizar Twint radica en su capacidad de obtener información de usuarios que no se encuentran disponibles mediante las APIs, así como de aquellos usuarios que han eliminado o bloqueado sus cuentas. Con esta herramienta, se podrá obtener una gran cantidad de datos en un corto período de tiempo, lo que permitirá construir una base de datos completa y representativa.

Además, Twint ofrece varias opciones de filtrado para especificar las palabras clave, el idioma, el rango de fechas, entre otros criterios, lo que permitirá obtener tweets relevantes y útiles para nuestro análisis de sentimientos. Con esta estrategia, se espera obtener una base de datos de alta calidad y representativa de la población objetivo, lo que aumentará la fiabilidad y validez de los resultados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
#Library for Scrappy Twitter
import twint

import nest_asyncio
nest_asyncio.apply()

import asyncio
asyncio.set_event_loop_policy(asyncio.WindowsSelectorEventLoopPolicy())
```

```python
# Configuración de Twint
c = twint.Config()
c.Search = "empty OR sadness OR enthusiasm OR neutral OR worry OR surprise OR love OR fun OR hate OR happiness OR boredom OR relief OR anger OR depressed OR anxious"
c.Lang = "en" # solo tweets en ingles
c.Limit = 3000 # límite de 3000 tweets
c.Store_csv = True
c.Output = "data_test_210303.csv" # archivo donde se guardarán los tweets
```

```python
# Ejecutar la búsqueda
twint.run.Search(c)
```

```python
# Convertir los resultados a un DataFrame de Pandas
data_test = pd.read_csv('data_test_210303.csv')
```
