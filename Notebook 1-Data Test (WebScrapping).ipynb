## PreparaciÃ³n de la data de prueba (Test Data)

Para llevar a cabo en este proyecto la recolecciÃ³n de datos, se utilizarÃ¡ la librerÃ­a Twint, una herramienta de web scrapping de Twitter que permite obtener tweets pÃºblicos sin necesidad de una API. La ventaja de utilizar Twint radica en su capacidad de obtener informaciÃ³n de usuarios que no se encuentran disponibles mediante las APIs, asÃ­ como de aquellos usuarios que han eliminado o bloqueado sus cuentas. Con esta herramienta, se podrÃ¡ obtener una gran cantidad de datos en un corto perÃ­odo de tiempo, lo que permitirÃ¡ construir una base de datos completa y representativa.

AdemÃ¡s, Twint ofrece varias opciones de filtrado para especificar las palabras clave, el idioma, el rango de fechas, entre otros criterios, lo que permitirÃ¡ obtener tweets relevantes y Ãºtiles para nuestro anÃ¡lisis de sentimientos. Con esta estrategia, se espera obtener una base de datos de alta calidad y representativa de la poblaciÃ³n objetivo, lo que aumentarÃ¡ la fiabilidad y validez de los resultados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
#Library for Scrappy Twitter
import twint

import nest_asyncio
nest_asyncio.apply()

import asyncio
asyncio.set_event_loop_policy(asyncio.WindowsSelectorEventLoopPolicy())
```

```python
# ConfiguraciÃ³n de Twint
c = twint.Config()
c.Search = "empty OR sadness OR enthusiasm OR neutral OR worry OR surprise OR love OR fun OR hate OR happiness OR boredom OR relief OR anger OR depressed OR anxious"
c.Lang = "en" # solo tweets en ingles
c.Limit = 3000 # lÃ­mite de 3000 tweets
c.Store_csv = True
c.Output = "data_test_210303.csv" # archivo donde se guardarÃ¡n los tweets
```

```python
# Ejecutar la bÃºsqueda
twint.run.Search(c)
```

```python
# Convertir los resultados a un DataFrame de Pandas
data_test = pd.read_csv('data_test_210303.csv')
```
